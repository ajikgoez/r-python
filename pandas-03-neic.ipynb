# Working with timeseries

```python
%pylab inline
import pandas as pd
```

```python
df = pd.read_csv('neic_new.csv', parse_dates=['time'], index_col=0)
df.head(10)
```

```python
df.info()
```

```python
df['year'] = df.index.year
```

```python
df.head()
```

```python
df.groupby('year').size().plot()
```

```python
df['month'] = df.index.month
```

```python
df.groupby('month').size().plot()
```

```python
df.groupby('month')['mag'].mean().plot()
```

```python
df.boxplot(column='mag', by='month')
```
